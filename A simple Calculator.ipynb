```python
import math
from tkinter import *
from tkinter import ttk
import numpy as np
```

```python
def Number(x):
    global cal_text
    cal_text = cal_text+str(x)
    text.set(cal_text)
def Clear():
    global cal_text
    cal_text = ""
    text.set("")
def ChangeSign():
    global cal_text
    cal_text=str((-1)*(float(cal_text)))
    text.set(cal_text)
def FindSquare():
    global cal_text
    cal_text=str(float(cal_text)*float(cal_text))
    text.set(cal_text)
def FindSquareRoot():
    global cal_text
    FindAnswer()
    cal_text=str(np.sqrt(float(cal_text)))
    text.set(cal_text)
def FindPercent():
    global cal_text
    FindAnswer()
    cal_text=str(float(cal_text)/(100))
    text.set(cal_text)
def FindAnswer():
    global cal_text
    result=str(eval(cal_text))
    text.set(result)
    cal_text=result
```

```python

#WINDOW PROPERTIES
cal=Tk()
cal.geometry("300x350")
cal.configure(bg="#00ffff")

#TITLE
title=Label(cal,text="Calculator",font=("sans-serif",20,"bold","underline"),fg="#800000",bg="#00ffff")
title.place(x=70,y=5)

#Output Box
cal_text=""
text=StringVar()
text=StringVar()
output=Entry(cal,width=25,font=("arial",15,"bold"),bg="LIME GREEN",justify=RIGHT,fg="wHITE",textvariable=text)
output.place(x=5,y=55)

#Create Buttons
btn1=Button(cal,text="1",width=4,height=2,activebackground="gold",command=lambda: Number(1))
btn1.place(x=17,y=115)
btn2=Button(cal,text="2",width=4,height=2,activebackground="gold",command=lambda: Number(2))
btn2.place(x=60,y=115)
btn3=Button(cal,text="3",width=4,height=2,activebackground="gold",command=lambda: Number(3))
btn3.place(x=103,y=115)
btn4=Button(cal,text="4",width=4,height=2,activebackground="gold",command=lambda: Number(4))
btn4.place(x=17,y=160)
btn5=Button(cal,text="5",width=4,height=2,activebackground="gold",command=lambda: Number(5))
btn5.place(x=60,y=160)
btn6=Button(cal,text="6",width=4,height=2,activebackground="gold",command=lambda: Number(6))
btn6.place(x=103,y=160)
btn7=Button(cal,text="7",width=4,height=2,activebackground="gold",command=lambda: Number(7))
btn7.place(x=17,y=203)
btn8=Button(cal,text="8",width=4,height=2,activebackground="gold",command=lambda: Number(8))
btn8.place(x=60,y=203)
btn9=Button(cal,text="9",width=4,height=2,activebackground="gold",command=lambda: Number(9))
btn9.place(x=103,y=203)
btn0=Button(cal,text="0",width=4,height=2,activebackground="gold",command=lambda: Number(0))
btn0.place(x=60,y=246)
plus=Button(cal,text="+",width=4,height=2,activebackground="gold",command=lambda: Number("+"))
plus.place(x=146,y=115)
minus=Button(cal,text="-",width=4,height=2,activebackground="gold",command=lambda: Number("-"))
minus.place(x=146,y=160)
mul=Button(cal,text="x",width=4,height=2,activebackground="gold",command=lambda: Number("*"))
mul.place(x=146,y=203)
div=Button(cal,text="/",width=4,height=2,activebackground="gold",command=lambda: Number("/"))
div.place(x=146,y=246)
dot=Button(cal,text=".",width=4,height=2,activebackground="gold",command=lambda: Number("."))
dot.place(x=17,y=246)
sign=Button(cal,text="+/-",width=4,height=2,activebackground="gold",command=ChangeSign)
sign.place(x=103,y=246)
mod=Button(cal,text="%",width=4,height=2,activebackground="gold",command=FindPercent)
mod.place(x=189,y=115)
rt=Button(cal,text="√",width=4,height=2,activebackground="gold",command=FindSquareRoot)
rt.place(x=232,y=115)
delete=Button(cal,text="x²",width=10,height=2,activebackground="Red",command=FindSquare)
delete.place(x=189,y=160)
enter=Button(cal,text="↩",width=10,height=2,activebackground="Green",command=FindAnswer)
enter.place(x=189,y=203)
clear=Button(cal,text="Clear",width=10,height=2,activebackground="light Blue",command=Clear)
clear.place(x=189,y=246)
```
